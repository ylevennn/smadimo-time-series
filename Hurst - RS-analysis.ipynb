# Вычисление показателя Херста с помощью R/S-анализа

В этом ноутбуке рассмотрим шаги и математические формулы для вычисления показателя Херста $ H $ с использованием R/S-анализа.

### Шаг 1: Разделение на интервалы

Разделите временной ряд на несколько подряд идущих неперекрывающихся интервалов длиной $ n $.

### Шаг 2: Центрирование временного ряда

Для каждого интервала $ j $ длиной $ n $ пусть у нас есть временной ряд $$ X ^ {(j)} = \{x_1^ {(j)}, x_2^ {(j)}, \ldots, x_n^ {(j)}\} $$.

1. **Вычисление среднего значения для интервала:**

   $$\bar{x}^ {(j)} = \frac{1}{n} \sum_{i=1}^{n} x_i ^ {(j)}$$

2. **Центрирование интервала:**

   $$Y_i ^ {(j)} = x_i^ {(j)} - \bar{x}^ {(j)}, \quad i = 1, 2, \ldots, n$$


### Шаг 3: Вычисление диапазона $ R(n) ^ {(j)} $ для каждого интервала

Для каждого интервала $ j $:

1. **Кумулятивные суммы внутри интервала:**

   $$Z_{i}^{(j)} = \sum_{k=1}^{i} Y_{k}^{(j)}, \quad i = 1, 2, \ldots, n$$

2. **Диапазон $ R(n)^ {(j)} $:**

   $$R(n)^ {(j)} = \max(Z_{i}^{(j)}) - \min(Z_{i}^{(j)}), \quad i = 1, 2, \ldots, n$$

### Шаг 4: Вычисление среднего квадратического отклонения $ S(n)^ {(j)} $ для каждого интервала

Для каждого интервала $ j $:

$$S(n)^ {(j)} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (Y_{i}^{(j)})^2}$$

### Шаг 5: Вычисление $ R/S(n) ^ {(j)}$ для каждого интервала

Для каждого интервала $ j $:

$$ R/S(n) ^ {(j)} = \frac{R(n) ^ {(j)}}{S(n)^ {(j)}}$$

### Шаг 6: Вычисление общего $R/S(n)$ для фиксированного размера интервала $n$

Для фиксированного размера интервала $ n $, общий $R/S(n)$ для временного ряда вычисляется как среднее значение $ R/S(n) ^ {(j)} $ по всем интервалам $j$.

$$ R/S(n) = \frac{1}{\frac{N}{n}} \sum_{i=1}^{\frac{N}{n}} R/S(n)^{(j)} $$

### Шаг 7: Вычисление множества $R/S(n)$ для всех размеров интервала $n$

Для каждого разумного размера интервала $n$ вычисляется величина $R/S(n)$, как функция от размера интервала $n$.

### Шаг 8: Логарифмическая линейная регрессия

Постройте зависимость $\log(R/S)$ от $\log(n)$:

1. **Линейная регрессия:**

   $$\log\left( \frac{R(n)}{S(n)} \right) = H \log(n) + C$$

   где $ H $ — показатель Херста (искомый нами), а $ C $ — константа.

2. **Вычисление показателя Херста $ H $:**

   Наклон прямой, полученной в результате регрессии, и является оценкой показателя Херста $ H $.
